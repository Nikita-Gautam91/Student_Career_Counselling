```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler,VectorIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
spark = SparkSession.builder.appName("project Linear Regression ").getOrCreate()
```

```python
dataframe = spark.read.csv("ad_pr.csv", header = True)
```

```python
type(dataframe)
```

```python
dataframe.show(100)
```

```python
dataframe = dataframe.drop('Serial No.')
```

```python
dataframe.show()
```

```python
dataframe.printSchema()
```

```python
from pyspark.sql.functions import col
new_data_frame = dataframe.select(*(col(c).cast("float").alias(c) for c in dataframe.columns))
```

```python
new_data_frame.printSchema()
```

```python
features = new_data_frame.drop('Chance of Admit')
```

```python
features.show()
```

```python
assembler = VectorAssembler( inputCols = features.columns, outputCol = "features")
```

```python
output = assembler.transform(new_data_frame)
```

```python
output.show()
```

```python
output = output.select("features","Chance of Admit")
```

```python
train_df,test_df = output.randomSplit([0.7,0.3])
```

```python
train_df.show()
test_df.show()
```

```python
lin_reg = LinearRegression(featuresCol ='features' ,labelCol = 'Chance of Admit')
linear_model = lin_reg.fit(train_df)
```

```python
print("Coefficients:" + str(linear_model.coefficients))
print("Intercept: " + str(linear_model.intercept))
```

```python
trainSummary = linear_model.summary
print("RMSE: %f" % trainSummary.rootMeanSquaredError)
print("MSE: %f" % trainSummary.meanSquaredError)
print("r2: %f" % trainSummary.r2)
```

```python
predictions = linear_model.transform(test_df)
predictions.select("prediction","Chance of Admit","features").show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

# R Squared (R2)
pred_evaluator = RegressionEvaluator(predictionCol="prediction",
                                      labelCol="Chance of Admit",
                                      metricName="r2")
r2 = pred_evaluator.evaluate(predictions)
print("R Squared (R2) on test data =", r2)

# Mean Squared Error (MSE)
pred_evaluator_mse = RegressionEvaluator(predictionCol="prediction",
                                          labelCol="Chance of Admit",
                                          metricName="mse")
mse = pred_evaluator_mse.evaluate(predictions)
print("Mean Squared Error (MSE) on test data =", mse)

# Root Mean Squared Error (RMSE)
pred_evaluator_rmse = RegressionEvaluator(predictionCol="prediction",
                                           labelCol="Chance of Admit",
                                           metricName="rmse")
rmse = pred_evaluator_rmse.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data =", rmse)
```

```python
from pyspark.ml.regression import RandomForestRegressor

# Define RandomForestRegressor
rf = RandomForestRegressor(featuresCol='features', labelCol='Chance of Admit')

# Fit the model
rf_model = rf.fit(train_df)

# Make predictions
rf_predictions = rf_model.transform(test_df)

# Show predictions
rf_predictions.select("prediction", "Chance of Admit", "features").show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

# Make predictions on the training data
train_predictions = rf_model.transform(train_df)

# Show predictions on the training data
train_predictions.select("prediction", "Chance of Admit", "features").show()

# Initialize the evaluator
evaluator = RegressionEvaluator(labelCol="Chance of Admit", predictionCol="prediction")

# Compute RMSE on training data
rmse_train = evaluator.setMetricName("rmse").evaluate(train_predictions)
print(f"Root Mean Squared Error (RMSE) on training data = {rmse_train}")

# Compute MSE on training data
mse_train = evaluator.setMetricName("mse").evaluate(train_predictions)
print(f"Mean Squared Error (MSE) on training data = {mse_train}")

# Compute R^2 on training data
r2_train = evaluator.setMetricName("r2").evaluate(train_predictions)
print(f"R Squared (R2) on training data = {r2_train}")

# To evaluate on test data, you can use similar code if not already evaluated
# Make predictions on the test data
test_predictions = rf_model.transform(test_df)

# Show predictions on the test data
test_predictions.select("prediction", "Chance of Admit", "features").show()

# Compute RMSE on test data
rmse_test = evaluator.setMetricName("rmse").evaluate(test_predictions)
print(f"Root Mean Squared Error (RMSE) on test data = {rmse_test}")

# Compute MSE on test data
mse_test = evaluator.setMetricName("mse").evaluate(test_predictions)
print(f"Mean Squared Error (MSE) on test data = {mse_test}")

# Compute R^2 on test data
r2_test = evaluator.setMetricName("r2").evaluate(test_predictions)
print(f"R Squared (R2) on test data = {r2_test}")
```

```python
from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator

# Define the Gradient Boosting Regressor
gbt = GBTRegressor(featuresCol='features', labelCol='Chance of Admit', maxIter=100)

# Fit the model on the training data
gbt_model = gbt.fit(train_df)

# Make predictions on the training data
train_predictions = gbt_model.transform(train_df)

# Make predictions on the test data
test_predictions = gbt_model.transform(test_df)

# Initialize the evaluator
evaluator = RegressionEvaluator(labelCol="Chance of Admit", predictionCol="prediction")
```

```python
# Compute evaluation metrics for training data
rmse_train = evaluator.setMetricName("rmse").evaluate(train_predictions)
mse_train = evaluator.setMetricName("mse").evaluate(train_predictions)
r2_train = evaluator.setMetricName("r2").evaluate(train_predictions)

print(f"Root Mean Squared Error (RMSE) on training data = {rmse_train}")
print(f"Mean Squared Error (MSE) on training data = {mse_train}")
print(f"R Squared (R2) on training data = {r2_train}")

# Compute evaluation metrics for test data
rmse_test = evaluator.setMetricName("rmse").evaluate(test_predictions)
mse_test = evaluator.setMetricName("mse").evaluate(test_predictions)
r2_test = evaluator.setMetricName("r2").evaluate(test_predictions)

print(f"Root Mean Squared Error (RMSE) on test data = {rmse_test}")
print(f"Mean Squared Error (MSE) on test data = {mse_test}")
print(f"R Squared (R2) on test data = {r2_test}")
```

```python
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator

# Define Decision Tree Regressor
dt = DecisionTreeRegressor(featuresCol='features', labelCol='Chance of Admit')

# Fit the model
dt_model = dt.fit(train_df)

# Make predictions
train_predictions = dt_model.transform(train_df)
test_predictions = dt_model.transform(test_df)

# Initialize the evaluators
mse_evaluator = RegressionEvaluator(labelCol="Chance of Admit", predictionCol="prediction", metricName="mse")
rmse_evaluator = RegressionEvaluator(labelCol="Chance of Admit", predictionCol="prediction", metricName="rmse")
r2_evaluator = RegressionEvaluator(labelCol="Chance of Admit", predictionCol="prediction", metricName="r2")

# Calculate MSE on training data
mse_train = mse_evaluator.evaluate(train_predictions)
# Calculate RMSE on training data
rmse_train = rmse_evaluator.evaluate(train_predictions)
# Calculate R2 on training data
r2_train = r2_evaluator.evaluate(train_predictions)

# Calculate MSE on test data
mse_test = mse_evaluator.evaluate(test_predictions)
# Calculate RMSE on test data
rmse_test = rmse_evaluator.evaluate(test_predictions)
# Calculate R2 on test data
r2_test = r2_evaluator.evaluate(test_predictions)

# Print the evaluation metrics
print("Decision Tree Regression Model Evaluation:")
print(f"MSE on training data: {mse_train}")
print(f"RMSE on training data: {rmse_train}")
print(f"R2 on training data: {r2_train}")
print(f"MSE on test data: {mse_test}")
print(f"RMSE on test data: {rmse_test}")
print(f"R2 on test data: {r2_test}")
```
